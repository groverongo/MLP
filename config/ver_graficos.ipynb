```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

p = pd.read_csv("../data/renuevo/perdidas.csv", header=None)
y_pred = pd.read_csv("../data/renuevo/renuevo.csv", header=None).to_numpy()

y_true = pd.read_csv("../res/training.csv", header=None).iloc[:,-24:].to_numpy()
print(y_pred.shape)
print(y_true.shape)

amax_y_pred = np.argmax(y_pred, axis=1)
amax_y_true = np.argmax(y_true, axis=1)
print(sum(np.equal(amax_y_pred, amax_y_true)))

plt.plot(p.index, p.values)
```
